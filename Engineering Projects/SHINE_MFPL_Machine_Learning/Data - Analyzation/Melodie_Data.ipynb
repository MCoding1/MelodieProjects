```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Data1..csv")
data.head()
data.corr()
```

```python
sns.heatmap(data.corr());
```

```python
#reliation .85 or higher is redundant -> do not use

# Increase the size of the heatmap.
plt.figure(figsize=(30, 10))
# Store heatmap object in a variable to easily access it when you want to include more features (such as title).
# Set the range of values to be displayed on the colormap from -1 to 1, and set the annotation to True to display the correlation values on the heatmap.
heatmap = sns.heatmap(data.corr(), vmin=-1, vmax=1, annot=True)
# Give a title to the heatmap. Pad defines the distance of the title from the top of the heatmap.
heatmap.set_title('Correlation Heatmap For Heart Data', fontdict={'fontsize':12}, pad=12);
```

```python
#triangle correlation heat map

plt.figure(figsize=(16, 6))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(data.corr(), dtype=np.bool))
heatmap = sns.heatmap(data.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```

```python
#split x and y data frames
from sklearn.model_selection import train_test_split
#x_train, x_test, y_train, y_test, y1_train, y1_test, y2_train, y2_test= train_test_split(x, y, y1, y2, random_state = 1234)
x_train, x_test, y_train, y_test= train_test_split(x, y, random_state = 1234)
```

training model

```python
from sklearn.linear_model import LinearRegression #for linear regression model
model = LinearRegression().fit(x_train, y_train)
```

```python
#estimated intercept
model.intercept_
```

```python
#estimated slope values/coefficients
model.coef_
```

evaluating

```python
#r^2, closer to 1 is better
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
#MAE calculation
mean_absolute_error(y_test, y_pred)
#expect predictions to be off by .05 +-
```
