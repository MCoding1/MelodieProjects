```python
#normalize

# copy the data
df_max_scaled = dataN.copy()

# apply normalization techniques
for column in df_max_scaled.columns:
    df_max_scaled[column] = df_max_scaled[column]  / df_max_scaled[column].abs().max()

# view normalized data
display(df_max_scaled)
```

```python
x00 = df_max_scaled[['X11', 'X12', 'X1', 'X2', 'X5', 'X13']]
x = x00.values
```

```python
response = 'X10'
y00 = df_max_scaled[[response]]
y = y00.values
```

```python
def function_tanh(input_dim, output_dim, width, activation):
#W[l] = np.random.randn(l-1,l)*10
  model = Sequential()
  model.add(Dense(units=width, activation='relu', input_dim=input_dim, kernel_initializer=initializers.RandomUniform(-1, 1),
                               bias_initializer=initializers.RandomUniform(-1, 1)))
  model.add(Dense(units=width, activation='relu', kernel_initializer=initializers.RandomUniform(-1, 1),
                               bias_initializer=initializers.RandomUniform(-1, 1)))
  model.add(Dense(units=output_dim, kernel_initializer=initializers.RandomUniform(-1, 1),
                               bias_initializer=initializers.RandomUniform(-1, 1))) #activation not here because linear is default
  # model.build()
  return model
```
